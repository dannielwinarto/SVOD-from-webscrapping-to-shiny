```python
import bs4 as bs
import urllib.request
import pandas as pd
import numpy as np
import time
from random import *
```

```python
imdb_id = pd.read_csv("imdb_id.csv", dtype = {0 : str, 1: str, 2:str})
imdb_id
```

```python
data_clean = imdb_id.dropna()
data_clean[:10]
```

```python
# the magic func:

def scrap_to_df(dataset):   
    count = 1
    id_keyword_df = pd.DataFrame()
    splitter = "- Plot keywords"
    for video_id, imdb_id in zip(dataset['source_video_id'].tolist(),dataset['imdb tittle id'].tolist()):
        url_link_1 = "http://www.imdb.com/title/"+ imdb_id + "/keywords"
        source = urllib.request.urlopen(url_link_1).read()
        soup = bs.BeautifulSoup(source, "html5lib") #for mac use "html5lib" tree parser, for windows use "lxml"
        keywords = []
        for each in soup.findAll('div', class_ = "sodatext"): # getting keywords
            temp = each.text.strip()
            keywords.append(temp)
        if not keywords: # if keywords null(exist in IMDB, keywords are empty), pass this iteration
            continue
        url_link_2 = "http://www.imdb.com/title/" + imdb_id
        source = urllib.request.urlopen(url_link_2).read()
        soup = bs.BeautifulSoup(source, "html5lib")
        try:
            rating = soup.find('div', class_ = "imdbRating").span.text # class = imdbRating, tag = "span"
            print (rating)
        except:
            rating = "no rating"
            print (rating)
        id_replicates = [video_id for x in (range(len(keywords)))]
        title_replicates = [soup.title.text.split(splitter,1)[0] for x in (range(len(keywords)))]
        ratings_replicates = [rating for x in (range(len(keywords)))]
        temp_list = list(zip(id_replicates, title_replicates, keywords, ratings_replicates))
        id_keyword_df = id_keyword_df.append(temp_list)
        random_pause = uniform(1, 10)
        print( "pause  = " + str(random_pause) + " count = " + str(count))
        count +=1
        time.sleep(random_pause)
    return id_keyword_df
```

```python
# scrap_to_df(data_clean.iloc[0:2])
output_df = scrap_to_df(data_clean[:10])
```
